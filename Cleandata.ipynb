```python
import pandas as pd
import numpy as np
import re
```

```python
#Importing the data
contracts_path = "contracts.csv"
details_path   = "contract_details.csv"

df_contracts = pd.read_csv(contracts_path)
df_details   = pd.read_csv(details_path)

print("contracts:", df_contracts.shape)
print("details:", df_details.shape)

df_contracts.head()
```

```python
merged_df = pd.merge(
    df_contracts,
    df_details,
    left_on='contract_code',
    right_on='contract_id',
    how='inner' # Used inner merge since both files have the same number of rows
)
```

```python
# --- 3. Clean Column Names (Remove Prefix) ---
# Identify columns with the prefix 'contractRecordsTO.' and rename them
prefix = 'contractRecordsTO.'
columns_to_rename = {
    col: col.replace(prefix, '')
    for col in merged_df.columns
    if prefix in col
}
merged_df.rename(columns=columns_to_rename, inplace=True)
```

```python
#Removing all the duplicated column
columns_to_drop = [
    'contractName',         
    'contractAmount',        
    'procurementCategoryId',
    'procurementMethodId',   
    'contractStatus',        
    'contractId'             
]
```

```python
merged_df.drop(columns=columns_to_drop, inplace=True, errors='ignore')

# --- 5. Save the Final Cleaned Data ---
cleaned_file_name = 'combined_contract_data.csv'
merged_df.to_csv(cleaned_file_name, index=False)
```
